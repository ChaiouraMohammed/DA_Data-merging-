```python
# Pie Chart
print(mergeData_cve_product['access_complexity'].value_counts())
access_complexity_counts = mergeData_cve_product['access_complexity'].value_counts()
plt.figure(figsize=(8, 6))
access_complexity_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#66c2a5', '#fc8d62', '#8da0cb'])
plt.title('Access Complexity Distribution')
plt.ylabel('')
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(8, 6))
sns.countplot(data=mergeData_cve_product, x='access_complexity', order=['LOW', 'MEDIUM', 'HIGH'], palette='viridis')
plt.title('Access Complexity Distribution')
plt.xlabel('Access Complexity')
plt.ylabel('Count')
plt.show()
```

```python


# Bar Plots
plt.figure(figsize=(18, 6))

# Bar plot for access_complexity
plt.subplot(1, 3, 1)
sns.countplot(data=mergeData_cve_product, x='access_complexity', order=['LOW', 'MEDIUM', 'HIGH'], palette='viridis')
plt.title('Access Complexity Distribution')
plt.xlabel('Access Complexity')
plt.ylabel('Count')

# Bar plot for impact_confidentiality
plt.subplot(1, 3, 2)
sns.countplot(data=mergeData_cve_product, x='impact_confidentiality', order=['NONE', 'PARTIAL', 'COMPLETE'], palette='viridis')
plt.title('Impact Confidentiality Distribution')
plt.xlabel('Impact Confidentiality')
plt.ylabel('Count')

# Bar plot for impact_integrity
plt.subplot(1, 3, 3)
sns.countplot(data=mergeData_cve_product, x='impact_integrity', order=['NONE', 'PARTIAL', 'COMPLETE'], palette='viridis')
plt.title('Impact Integrity Distribution')
plt.xlabel('Impact Integrity')
plt.ylabel('Count')

# Display the plots
plt.tight_layout()
plt.show()

# Pie Charts
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Pie chart for access_complexity
access_complexity_counts = mergeData_cve_product['access_complexity'].value_counts()
access_complexity_counts.plot.pie(ax=axes[0], autopct='%1.1f%%', startangle=90, colors=['#66c2a5', '#fc8d62', '#8da0cb'])
axes[0].set_title('Access Complexity Distribution')
axes[0].set_ylabel('')

# Pie chart for impact_confidentiality
impact_confidentiality_counts = mergeData_cve_product['impact_confidentiality'].value_counts()
impact_confidentiality_counts.plot.pie(ax=axes[1], autopct='%1.1f%%', startangle=90, colors=['#66c2a5', '#fc8d62', '#8da0cb'])
axes[1].set_title('Impact Confidentiality Distribution')
axes[1].set_ylabel('')

# Pie chart for impact_integrity
impact_integrity_counts = mergeData_cve_product['impact_integrity'].value_counts()
impact_integrity_counts.plot.pie(ax=axes[2], autopct='%1.1f%%', startangle=90, colors=['#66c2a5', '#fc8d62', '#8da0cb'])
axes[2].set_title('Impact Integrity Distribution')
axes[2].set_ylabel('')

# Display the plots
plt.tight_layout()
plt.show()
```
